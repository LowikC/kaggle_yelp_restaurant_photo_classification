```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
%matplotlib inline
```

```python
train_df = pd.read_csv('../data/models/multilabel/caffenet/caffe.INFO.train')
```

```python
train_df.head(5)
```

```python
val_df = pd.read_csv('../data/models/multilabel/caffenet/caffe.INFO.test')
```

```python
val_df.head(5)
```

```python
host = host_subplot(111, axes_class=AA.Axes)
plt.subplots_adjust(right=0.75)

par1 = host.twinx()

host.set_xlabel("Iterations")
host.set_ylabel("Loss")
par1.set_ylabel("Learning rate")

p1, = host.plot(train_df.NumIters, train_df.loss, label="Training Loss")
p1, = host.plot(val_df.NumIters, val_df.loss, linewidth=2, label="Validation Loss")
p2, = par1.plot(train_df.NumIters, train_df.LearningRate, label="Learning Rate")

host.legend(loc='upper left')
host.set_title("CaffeNet - MultiLabel")

host.axis["left"].label.set_color(p1.get_color())
par1.axis["right"].label.set_color(p2.get_color())

plt.draw()

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```
